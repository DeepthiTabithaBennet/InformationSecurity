```python
# Written by Deepthi Tabitha Bennet

import numpy as np
from sympy import Matrix
```

```python
def HillEncrypt(message, key):

    encryptedMessage = ""

    values = [[0], [0], [0]]
    for j in range(0, 3):
        values[j][0] = (ord(message[j]) % 65)

    # Create the vectors perform matrix multiplication
    vector = np.matrix(values)
    vector = (key * vector) % 26

    for j in range(0, 3):
        encryptedMessage += chr(vector.item(j) + 65)

    return(encryptedMessage)
```

```python
def HillDecrypt(message, key):

    decryptedMessage = ""

    # Get the decrypt key
    key = Matrix(key)
    key = key.inv_mod(26)
    key = key.tolist()

    values = [[0], [0], [0]]
    for j in range(0, 3):
        values[j][0] = (ord(message[j]) % 65)

    # Create the vectors perform matrix multiplication
    vector = np.matrix(values)
    vector = (key * vector) % 26

    for j in range(0, 3):
        decryptedMessage += chr(vector.item(j) + 65)

    return(decryptedMessage)
```

```python
key = np.matrix([[6, 24, 1], [13, 16, 10], [20, 17, 15]])

message = input("Enter your message : ").replace(" ", '')
message = message.upper()

plaintext_encrypted = HillEncrypt(message, key)
print("Encrypted Text :", plaintext_encrypted)

ciphertext_decrypted = HillDecrypt(plaintext_encrypted, key)
print("Decrypted Text :", ciphertext_decrypted)
```
