```python
# Written by Deepthi Tabitha Bennet

string = input("input string : ")
key = input("input keyword : ").replace(' ', '') # remove all spaces

alphabet = "abcdefghijklmnopqrstuvwxyz"
text = []

# remove duplicates from the key
for i in key:
    if i not in text:
        text.append(i)  # append only unique characters

# fill the rest with missing letters in alphabet
for i in alphabet:
    if i not in text:
        text.append(i)

print(text)
```

```python
def encryption(string):
    encrypted_string = ""

    for char in string:

        if char.isalpha():
            char_index = alphabet.index(char.lower())
            encrypted_char = text[char_index]
            encrypted_string += encrypted_char

        else:
            encrypted_string += char

    print("Encrypted : ", encrypted_string)
    return encrypted_string
```

```python
def decryption(encrypted_string):
    decrypted_string = ""

    for char in encrypted_string:

        if char.isalpha():
            char_index = text.index(char.lower())
            decrypted_char = alphabet[char_index]
            decrypted_string += decrypted_char

        else:
            decrypted_string += char

    print("Decrypted : ", decrypted_string)
    return decrypted_string
```

```python
enc_msg = encryption(string)
dec_msg = decryption(enc_msg)
```
