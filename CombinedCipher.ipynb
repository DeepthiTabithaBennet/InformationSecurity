## **Caesar Cipher**

```python
def caesar_encrypt(plaintext, shift):
    plaintext_encrypted = ''

    for letter in plaintext:

        if letter.isdigit():
            plaintext_encrypted += chr((ord(letter) + shift - ord('0')) % 10 + ord('0'))  # We perform %10 as there are 10 digits from 0 to 9

        elif letter.isupper():
            plaintext_encrypted += chr((ord(letter) + shift - ord('A')) % 26 + ord('A'))  # We perform %26 as there are 26 characters from A to Z

        elif letter.islower():
            plaintext_encrypted += chr((ord(letter) + shift - ord('a')) % 26 + ord('a'))  # We perform %26 as there are 26 digits from a to z

        else:
            plaintext_encrypted += letter

    print("Caesar Cipher Encrypted Text: ", plaintext_encrypted)
    return(plaintext_encrypted)
```

```python
def caesar_decrypt(plaintext, shift):
    ciphertext_decrypted = ''

    for letter in plaintext:

        if letter.isdigit():
            ciphertext_decrypted += chr((ord(letter) - shift - ord('0')) % 10 + ord('0'))  # We perform %10 as there are 10 digits from 0 to 9

        elif letter.isupper():
            ciphertext_decrypted += chr((ord(letter) - shift - ord('A')) % 26 + ord('A'))  # We perform %26 as there are 26 characters from A to Z

        elif letter.islower():
            ciphertext_decrypted += chr((ord(letter) - shift - ord('a')) % 26 + ord('a'))  # We perform %26 as there are 26 digits from a to z

        else:
            ciphertext_decrypted += letter

    print("Caesar Cipher Decrypted Text: ", ciphertext_decrypted)
    return(ciphertext_decrypted)
```

## **Vigenere Cipher**

```python
def generateKey(plaintext, key):

    key = list(key)

    if len(plaintext) == len(key):
        return(key)

    else:
        for i in range(len(plaintext) - len(key)):
            key.append(key[i % len(key)])

    return("" . join(key))
```

```python
def vigenere_encrypt(plaintext, key):

    plaintext_encrypted = []

    for i in range(len(plaintext)):
        x = (ord(plaintext[i]) + ord(key[i])) % 26
        x += ord('A')
        plaintext_encrypted.append(chr(x))

    print("Vigenere Cipher Encrypted Text: ", ("" . join(plaintext_encrypted)))
    return("" . join(plaintext_encrypted))
```

```python
def vigenere_decrypt(plaintext_encrypted, key):

    orig_text = []

    for i in range(len(plaintext_encrypted)):
        x = (ord(plaintext_encrypted[i]) - ord(key[i]) + 26) % 26
        x += ord('A')
        orig_text.append(chr(x))

    print("Vigenere Cipher Decrypted Text: ", ("" . join(orig_text)))
    return("" . join(orig_text))
```

```python
key = generateKey(plaintext, key)

print("Plaintext: ", plaintext)

caesar_enc = caesar_encrypt(plaintext, shift)
vigenere_enc = vigenere_encrypt(caesar_enc, key)

vigenere_dec = vigenere_decrypt(vigenere_enc, key)
caesar_dec = caesar_decrypt(vigenere_dec, shift)
```
