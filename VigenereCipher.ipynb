```python
# Written by Deepthi Tabitha Bennet

plaintext = input("Enter your message: ")
plaintext = plaintext.replace(" ", "")
plaintext = plaintext.upper()

key = input("Enter the keyword: ")
key = key.replace(" ", "")
key = key.upper()
```

# **From Scratch**

```python
def generateKey(plaintext, key):

    if len(plaintext) != len(key):
        for i in range(len(plaintext) - len(key)):
            key += key[i % len(key)]

    return key
```

```python
def encryption(plaintext, key):

    plaintext_encrypted = ""

    for i in range(len(plaintext)):
        x = (ord(plaintext[i]) + ord(key[i])) % 26 + ord('A')
        plaintext_encrypted += chr(x)

    return plaintext_encrypted
```

```python
def decryption(plaintext_encrypted, key):

    orig_text = ''

    for i in range(len(plaintext_encrypted)):
        x = (ord(plaintext_encrypted[i]) - ord(key[i]) + 26) % 26 + ord('A')
        orig_text += chr(x)

    return orig_text
```

```python
key = generateKey(plaintext, key)
plaintext_encrypted = encryption(plaintext,key)
ciphertext_decrypted = decryption(plaintext_encrypted, key)

print("Encrypted message : ", plaintext_encrypted)
print("Decrypted message : ", ciphertext_decrypted)
```
