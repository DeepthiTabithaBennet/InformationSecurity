```python
# Written by Deepthi Tabitha Bennet

plaintext = input("Enter your message: ").replace(" ", '')
plaintext = plaintext.lower()
```

# **From Scratch**

```python
def encryption(text):
    N = ord('z') + ord('a')
    ans = ''

    for s in text:
        ans += chr(N - ord(s))

    return ans
```

```python
def decryption(text):  # same procedure as encrypt function
    N = ord('z') + ord('a')
    ans = ''

    for s in text:
        ans += chr(N - ord(s))

    return ans
```

```python
plaintext_encrypted = encryption(plaintext)
print("Encrypted Text:", plaintext_encrypted)

ciphertext_decrypted = decryption(plaintext_encrypted)
print("Decrypted Text:", ciphertext_decrypted)
```
