# **From Scratch**

```python
def generate_key_table(key):

    # Creating a key table
    key = key.replace(" ", "").upper()  # Remove spaces and convert to uppercase
    key_table = ""
    alphabet = "ABCDEFGHIKLMNOPQRSTUVWXYZ"  # Excluding 'J' for Playfair cipher

    for char in key:
        if char not in key_table
            key_table += char

    for char in alphabet:
        if char not in key_table:
            key_table += char

    return key_table
```

```python
def generate_pairs(text):

    # Generating pairs of characters from the text
    text = text.replace(" ", "").upper()  # Remove spaces and convert to uppercase
    pairs = []
    i = 0

    while i < len(text):

        if i == len(text) - 1 or text[i] == text[i + 1]:
            pairs.append(text[i] + 'X')
            i += 1

        else:
            pairs.append(text[i] + text[i + 1])
            i += 2

    return pairs
```

```python
def encrypt(plaintext, key):

    key_table = generate_key_table(key)
    pairs = generate_pairs(plaintext)
    ciphertext = ""

    for pair in pairs:
        row1, col1 = divmod(key_table.index(pair[0]), 5)
        row2, col2 = divmod(key_table.index(pair[1]), 5)

        if row1 == row2:
            ciphertext += key_table[row1 * 5 + (col1 + 1) % 5]
            ciphertext += key_table[row2 * 5 + (col2 + 1) % 5]

        elif col1 == col2:
            ciphertext += key_table[((row1 + 1) % 5) * 5 + col1]
            ciphertext += key_table[((row2 + 1) % 5) * 5 + col2]

        else:
            ciphertext += key_table[row1 * 5 + col2]
            ciphertext += key_table[row2 * 5 + col1]

    return ciphertext
```

```python
def decrypt(ciphertext, key):

    key_table = generate_key_table(key)
    pairs = generate_pairs(ciphertext)
    plaintext = ""

    for pair in pairs:
        row1, col1 = divmod(key_table.index(pair[0]), 5)
        row2, col2 = divmod(key_table.index(pair[1]), 5)

        if row1 == row2:
            plaintext += key_table[row1 * 5 + (col1 - 1) % 5]
            plaintext += key_table[row2 * 5 + (col2 - 1) % 5]

        elif col1 == col2:
            plaintext += key_table[((row1 - 1) % 5) * 5 + col1]
            plaintext += key_table[((row2 - 1) % 5) * 5 + col2]

        else:
            plaintext += key_table[row1 * 5 + col2]
            plaintext += key_table[row2 * 5 + col1]

    return plaintext
```

```python
ciphertext = encrypt(plaintext, key)
print("Encrypted Text : ", ciphertext)

decrypted_text = decrypt(ciphertext, key)
print("Decrypted Text : ", decrypted_text)
```
