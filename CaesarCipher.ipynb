```python
# Written by Deepthi Tabitha Bennet

plaintext = input("Enter plaintext to be Encrypted: ")
shift = int(input("Enter the Shift Direction (+/-), and Shift Value (int): "))
```

# **From Scratch**

```python
plaintext_encrypted = ''

# ord() -> character to ASCII
# chr() -> ASCII to character

for letter in plaintext:

    if letter.isdigit():  # Remove these two lines to match In-Built Function
        plaintext_encrypted += chr((ord(letter) + shift - ord('0')) % 10 + ord('0'))  # We perform %10 as there are 10 digits from 0 to 9

    elif letter.isupper():
        plaintext_encrypted += chr((ord(letter) + shift - ord('A')) % 26 + ord('A'))  # We perform %26 as there are 26 characters from A to Z

    elif letter.islower():
        plaintext_encrypted += chr((ord(letter) + shift - ord('a')) % 26 + ord('a'))  # We perform %26 as there are 26 digits from a to z

    else:
        plaintext_encrypted += letter

print("Encrypted Text: ", plaintext_encrypted)
```

```python
ciphertext_decrypted = ''

# ord() -> character to ASCII
# chr() -> ASCII to character

for letter in plaintext_encrypted:

    if letter.isdigit():  # Remove these two lines to match In-Built Function
        ciphertext_decrypted += chr((ord(letter) - shift - ord('0')) % 10 + ord('0'))  # We perform %10 as there are 10 digits from 0 to 9

    elif letter.isupper():
        ciphertext_decrypted += chr((ord(letter) - shift - ord('A')) % 26 + ord('A'))  # We perform %26 as there are 26 characters from A to Z

    elif letter.islower():
        ciphertext_decrypted += chr((ord(letter) - shift - ord('a')) % 26 + ord('a'))  # We perform %26 as there are 26 digits from a to z

    else:
        ciphertext_decrypted += letter

print("Decrypted Text: ", ciphertext_decrypted)
```

# **In-Built Modules - caesarcipher**

```python
cipher = CaesarCipher(plaintext, offset=shift)
print("Encrypted Text: ", cipher.encoded)
print("Decrypted Text: ", cipher.decoded)
```
